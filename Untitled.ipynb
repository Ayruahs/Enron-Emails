```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import ceil

filename = 'emails.csv'
dataset = pd.read_csv(filename)

emails = dataset['message']

# Functions to return name of sender and content of email

def returnName(email):
    lines = email.split('\n')
    for i in range(len(lines)):
        if 'X-Origin:' in lines[i]:
            name = lines[i].split()[1]
            return name.title()

    
def returnContent(email):
    lines = email.split('\n')
    for i in range(len(lines)):
        if 'X-FileName:' in lines[i]:
            break
        else:
            i = -1
    content = '\n'.join(lines[i+2:])  
    return content

# Function to return size of string
def returnSize(s):
    return len(s.encode('utf-8'))

# Return Date
def returnDate(email):
    message = email.split('\n')
    date = message[1][5:]
    return date

contents = []
names = []
stringDates = []
bytesizes = []
for email in emails:
    contents.append(returnContent(email))
    names.append(returnName(email))
    stringDates.append(returnDate(email))
    bytesizes.append(returnSize(email))    

dataset['names'] = pd.Series(names)
dataset['Email Content'] = pd.Series(contents)
    
dates = pd.to_datetime(stringDates, infer_datetime_format=True)

#dataset['hour'] = pd.Series((dates.hour*3600+dates.minute*60+dates.second)/3600)
#dataset['size'] = pd.Series(bytesizes)
```
