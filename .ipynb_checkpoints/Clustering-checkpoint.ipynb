```python
# Plot size of email contents of users (in bytes) and hour of day they sent the email and 
# look for anomalous behavior

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import ceil

filename = 'emails.csv'
dataset = pd.read_csv(filename, nrows=10000)
#dataset = pd.read_csv(filename)

print(dataset.shape)
print(dataset.head())
print(dataset.describe())
```

```python
emails = dataset['message']

# Functions to return name of sender and content of email

def returnName(email):
    lines = email.split('\n')
    for i in range(len(lines)):
        if 'X-Origin:' in lines[i]:
            name = lines[i].split()[1]
            return name.title()

    
def returnContent(email):
    lines = email.split('\n')
    for i in range(len(lines)):
        if 'X-FileName:' in lines[i]:
            break
        else:
            i = -1
    content = '\n'.join(lines[i+2:])  
    return content

# Function to return size of string
def returnSize(s):
    return len(s.encode('utf-8'))

# Return Date
def returnDate(email):
    message = email.split('\n')
    date = message[1][5:]
    return date
```

```python
contents = []
names = []
stringDates = []
bytesizes = []
for email in emails:
    contents.append(returnContent(email))
    names.append(returnName(email))
    stringDates.append(returnDate(email))
    bytesizes.append(returnSize(email))    

dataset['names'] = pd.Series(names)
dataset['Email Content'] = pd.Series(contents)
    
dates = pd.to_datetime(stringDates, infer_datetime_format=True)

dataset['hour'] = pd.Series((dates.hour*3600+dates.minute*60+dates.second)/3600)
dataset['size'] = pd.Series(bytesizes)

#dataset.index.names = ['names']
#dataset.index = names

print(dataset.head())
```

```python
newNames = list(set(names))
#fig, axes = plt.subplots(ceil(len(newNames)//2), 2, figsize=[24,500])#, dpi=180)
fig, axes = plt.subplots(ceil(len(newNames)//2), 2, figsize=[24,24])#, dpi=180)
for i in range(len(newNames)):
    nameHours = dataset.loc[(dataset["names"] == newNames[i]), 'hour'].values
    nameBytes = dataset.loc[(dataset["names"] == newNames[i]), 'size'].values 
    
    axes[i//2][i%2].scatter(nameHours, nameBytes, linewidths=0, marker='.', color='blue')
    axes[i//2][i%2].set_xlabel(newNames[i])
    axes[i//2][i%2].set_xlim([0,24])
    
plt.show()
```
